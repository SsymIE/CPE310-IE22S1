**Loop control** <br>

**break and continue** - alter loop fow dynamically. 'Break' terminate the entire loop immediately. 'Continue' skips only the remaining code in the current iteration and jumps to to the next cycle

```python
for i in range (1, 6):
  if i == 3:
    continue
  print(i)
```

```python
for i in range(1, 6):
  if i == 3:
     break
  print(i)
```
