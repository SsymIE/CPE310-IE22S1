```python
# 1. Create a list

fruits = ["apple", "banana", "cherry", "orange"]
print("Original list:", fruits)
```

```python
# 2. Access elements


print ("First fruit:", fruits[0])
print ("Last fruit:", fruits[-1])
```

```python
# 3. Add elements

fruits.append("mango")                    # add at the end
fruits.insert(1, "grape")                 #  add at specific position
print("After adding fruits:", fruits)
```

```python
# 4. Remove elements

fruits.remove("banana")           # remove by value
popped = fruits.pop()             # remove last item
print("After removing fruits:", fruits)
print("Popped item:", popped)
```

```python
# 5. Check if an item exist

if "apple" in fruits:
    print("Apple is in the list!")
```

```python
# 6. Loop through the list

print("All fruits in the list:")
for fruit in fruits:
    print("-", fruit)
```

```python
# 7. Lenght of the list

print("Total number of fruits", len(fruits))
```

```python
# 8. Sorting the list

fruits.sort()                 #Alphabetical order
print("Sorted list:", fruits)
```

```python
# 9. Reversing the list

fruits.reverse()
print("Reversed list:", fruits)
```
